```python
import pandas as pd 

df = pd.read_csv("brennan_johnson_grouped.csv")
pd.set_option('display.max_rows', None)  
pd.set_option('display.max_columns', None)  

# تأكد أن أسماء الأعمدة مكتوبة بشكل صحيح
columns_to_update = ['games_played', 'starts', 'mins', 'goals', 'shots', 
                     'shots_on_target', 'g+a', 'non_pens_goals', 'xg', 
                     'npxg', 'npxg+xa', 'takes_on_attemped', 'takes_on_succeeded']

values_to_assign = [33, 23, 2117 + 64, 14, 79, 26, 17, 14, 12.05,12.05, 12.05 + 3.6, 42, 20]

# تأكد أن عدد القيم يطابق عدد الأعمدة
if len(columns_to_update) == len(values_to_assign):
    df.loc[5, columns_to_update] = values_to_assign
else:
    print("❌ عدد القيم لا يطابق عدد الأعمدة!")

df.drop(columns=['Unnamed: 0'], inplace=True)
df.to_csv("brennan_johnson_grouped.csv", index=False)

# تحقق من التحديث
df
```

```python
import pandas as pd

df['goals_per90'] = df['goals'] / df['mins'] * 90
df['assists_per90'] = df['assists'] / df['mins'] * 90
df['goals_and_assists_per90'] = df['g+a'] / df['mins'] * 90
df['non_pens_goals_per90'] = df['non_pens_goals'] / df['mins'] * 90
df['xg_per90'] = df['xg'] / df['mins'] * 90
df['xa_per90'] = df['xa'] / df['mins'] * 90
df['xg_xa_per90'] = df['npxg+xa'] / df['mins'] * 90
df['npxg_per90'] = df['npxg'] / df['mins'] * 90
df['npxg_xa_per90'] = df['npxg+xa'] / df['mins'] * 90
df['g+a_per90'] = df['g+a'] / df['mins'] * 90

df.to_csv("brennan_johnson_grouped.csv")

df
```
