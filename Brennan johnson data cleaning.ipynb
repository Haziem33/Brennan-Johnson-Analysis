```python
import pandas as pd 
df = pd.read_csv("brennan_johnson_stats.csv")

df.columns = df.iloc[0]  
df = df[1:].reset_index(drop=True)  

df.columns = df.columns.str.strip()
df.drop("Matches", axis=1, inplace=True)

df.to_csv("brennan_johnson_stats_cleaned.csv")

df.head()



```

```python
df = df.iloc[:20]  # الاحتفاظ بالصفوف من 0 إلى 19 فقط
df = df.reset_index(drop=True)



print(df[df.columns[0]])
```

```python

df.drop(df[df["MP"] == '0'].index, inplace=True)


df
```

```python
df.columns = ['season', 'age', 'squad', 'country', 'comp', 'games_played','starts', 'mins','90s', 'goals', 'assists', 'g+a', 'non_pens_goals', 'pens', 'pkatt','yellow_card' ,"red_card","xg",'npxg','xa','npxg+xa','progC','progP','progR', 'goals_per90', 'assists_per90', 'goals_and_assists_per90', 'non_pens_goals_per90', 'goals_assists_non_pens_per90','xg_per90', 'xa_per90', 'xg_xa_per90', 'npxg_per90', 'npxg_xa_per90']

df.drop(columns=["90s",'pkatt'], inplace=True)



print(df.columns)

df.info()
```

```python

df['season'] = df['season'].astype(str)
df['age'] = df['age'].astype('Int8')   
df['games_played'] = df['games_played'].astype(float)
df['starts'] = df['starts'].astype(float)
df['mins'] = df['mins'].astype(float)
df['goals'] = df['goals'].astype(float)
df['assists'] = df['assists'].astype(float)
df['g+a'] = df['g+a'].astype(float)
df['non_pens_goals'] = df['non_pens_goals'].astype(float)
df['pens'] = df['pens'].astype(float)
df['yellow_card'] = df['yellow_card'].astype(float)
df['red_card'] = df['red_card'].astype(float)
df['xg'] = df['xg'].astype(float)
df['npxg'] = df['npxg'].astype(float)
df['xa'] = df['xa'].astype(float)
df['npxg+xa'] = df['npxg+xa'].astype(float)
df['progC'] = df['progC'].astype(float)
df['progP'] = df['progP'].astype(float)
df['progR'] = df['progR'].astype(float)
df['goals_per90'] = df['goals_per90'].astype(float)
df['assists_per90'] = df['assists_per90'].astype(float)
df['goals_and_assists_per90'] = df['goals_and_assists_per90'].astype(float)
df['non_pens_goals_per90'] = df['non_pens_goals_per90'].astype(float)
df['goals_assists_non_pens_per90'] = df['goals_assists_non_pens_per90'].astype(float)
df['xg_per90'] = df['xg_per90'].astype(float)
df['xa_per90'] = df['xa_per90'].astype(float)
df['xg_xa_per90'] = df['xg_xa_per90'].astype(float)
df['npxg_per90'] = df['npxg_per90'].astype(float)
df['npxg_xa_per90'] = df['npxg_xa_per90'].astype(float)

df.info()
```

```python
df['npxg'].fillna(df['xg'], inplace=True)

df['xa_per90'] = df['xa'] / df['mins'] * 90

df.fillna(df.select_dtypes(include=['object']).mode().iloc[0], inplace=True)


df.loc[df['games_played'] > 5, df.select_dtypes(include=['float']).columns] = \
df.loc[df['games_played'] > 5, df.select_dtypes(include=['float']).columns].fillna(df.select_dtypes(include=['float']).std())

df.fillna(0.7, inplace=True)



df.info()
```
